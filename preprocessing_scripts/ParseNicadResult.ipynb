```python
import pandas as pd
import xml.etree.ElementTree as ET
import os
```

# Parse Nicad XML to DataFrame

```python
def parse_NiCad_xml(file):
    root = ET.parse(file).getroot()
    NiCad = []
    repoName = file.split('_')[0].split('/')[-1]
    for child in root:
        if child.tag == 'systeminfo':        
            granularity = child.get('granularity')
        for source in child:
            NiCad.append({'repoName':repoName,
                          'classid': child.get('classid'),
                          'file': source.get('file'), 
                          'startline': source.get('startline')
                     , 'endline': source.get('endline'), 
                          'pcid': source.get('pcid'), 
                          'similarity':child.get('similarity')
                         ,'granularity':granularity})
    return pd.DataFrame(NiCad)
```
